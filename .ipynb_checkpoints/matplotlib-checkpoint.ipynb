```python
import numpy as np
t = np.arange(0 , 2*np.pi, 0.01)
    
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t))
plt.grid()
```

```python
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t))
plt.grid()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('sin')
plt.show()
```

```python
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t) , label = 'sin')
plt.plot(t , np.cos(t) , label = 'cos')
plt.grid()
plt.legend()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```

위에 그래프코드 해석
grid는 평면좌표 격자무늬 표시
legend는 왼쪽 하단 sin은 무슨 색 cos 무슨 색 표시

```python
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t) , 'k' ,lw = 3 , label = 'sin')
plt.plot(t , np.cos(t) , 'm' , lw = 6 ,label = 'cos')
plt.grid()
plt.legend()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```

lw는 선 굵기 옵션이고
blue = 'b'
red = 'r'
green = 'g'
cyan = 'c'
magenta = 'm'
yellow = 'y'
black = 'k'

```python
col_names = ['스트레스' , '스트레스남학생' , '스트레스여학생' , '우울감경험률' , 
            '우울남학생' , '우울여학생' , '자살생각율' , '자살남학생' , '자살여학생']
raw_data = pd.read_excel("./data/teenage_mental.xls",
                        header = 1,
                        usecols = 'C:K',
                        names = col_names)
raw_data.loc[1] = 100. - raw_data.loc[0]
raw_data['응답'] = ['그렇다','아니다']
raw_data.set_index('응답', drop = True , inplace = True)

raw_data
```
