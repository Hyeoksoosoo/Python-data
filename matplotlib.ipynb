```python
import numpy as np
t = np.arange(0 , 2*np.pi, 0.01)
    
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t))
plt.grid()
```

```python
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t))
plt.grid()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('sin')
plt.show()
```

```python
plt.figure(figsize = (10 , 6))
plt.plot(t , np.sin(t) , label = 'sin')
plt.plot(t , np.cos(t) , label = 'cos')
plt.grid()
plt.legend()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```
